### 4.1. Variables quantitatives

**Imputation par la médiane, la moyenne et le mode**

```python
from sklearn.impute import MissingIndicator, KNNImputer,SimpleImputer
```

- L'imputation par la moyenne, la médiane et le mode sont des méthodes simples pour remplacer les valeurs manquantes dans une variable.

- L'imputation par la moyenne consiste à remplacer les valeurs manquantes par la moyenne des valeurs non manquantes dans la variable. Cette méthode est utile pour les variables continues et symétriques.

- L'imputation par la médiane consiste à remplacer les valeurs manquantes par la médiane des valeurs non manquantes dans la variable. Cette méthode est utile pour les variables continues qui ont une distribution asymétrique ou pour les variables ordinales.

- L'imputation par le mode consiste à remplacer les valeurs manquantes par le mode, c'est-à-dire la valeur la plus fréquente, dans la variable. Cette méthode est utile pour les variables catégorielles ou nominales.

Ces méthodes sont simples à mettre en œuvre et peuvent être efficaces lorsque les données manquantes sont MCAR (Missing Completely At Random), c'est-à-dire que les valeurs manquantes sont indépendantes des autres variables. Toutefois, elles peuvent introduire un biais dans les données si les valeurs manquantes sont liées à d'autres variables dans les données.

### 4.2. Variables catégorielles

- **Label Encoding** : Le Label Encoding est une technique de transformation de variables catégorielles en valeurs numériques ordinales. Chaque catégorie possible de la variable est remplacée par un entier unique, en fonction de l'ordre dans lequel les catégories apparaissent dans la variable.

En général, le Label Encoding est une technique simple et rapide pour encoder des variables catégorielles en valeurs numériques ordinales, mais il est important de bien comprendre ses limites et d'utiliser d'autres techniques d'encodage pour des analyses plus avancées ou pour des algorithmes de machine learning plus complexes.

```python
import pandas as pd

# Fonction pour identifier les colonnes catégorielles
def get_categorical_columns(df):
    return df.select_dtypes(include=['object']).columns

# Fonction pour encoder toutes les colonnes catégorielles
def encode_all_categories(df):
    encoded_df = df.copy()  # Copie du DataFrame pour éviter de modifier l'original
    mapping_dict = {}  # Dictionnaire pour stocker les mappings

    # Identifier les colonnes catégorielles
    categorical_columns = get_categorical_columns(df)

    # Parcours des colonnes catégorielles
    for col in categorical_columns:
        # Obtenir les valeurs uniques (en ignorant les NA)
        unique_values = df[col].dropna().unique()

        # Créer un mapping de 0 à n pour chaque modalité
        mapping = {value: idx for idx, value in enumerate(unique_values)}

        # Appliquer le mapping à la colonne
        encoded_df[col] = df[col].map(mapping)

        # Stocker le mapping pour référence future
        mapping_dict[col] = mapping

    return encoded_df, mapping_dict

# Encodage des variables catégorielles
encoded_data, train_mappings = encode_all_categories(df)
```

```python
# Affichage des résultats
print(f"Colonnes catégorielles encodées : {len(get_categorical_columns(df))}")
print(f"Colonnes encodées : {get_categorical_columns(df).tolist()}\n")
```

# IV/ Choix des modèles

## 1. Regression linéaire multiple

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
model = LinearRegression()

# Entraîner le modèle
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, median_absolute_error, explained_variance_score,r2_score,mean_squared_error

# Prédiction sur la base d'apprentissage et la base de test
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Calcul des métriques sur la base d'apprentissage et la base de test
train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
train_mae = mean_absolute_error(y_train, y_train_pred)
test_mae = mean_absolute_error(y_test, y_test_pred)
train_medae = median_absolute_error(y_train, y_train_pred)
test_medae = median_absolute_error(y_test, y_test_pred)
train_r2 = r2_score(y_train, y_train_pred)
test_r2 = r2_score(y_test, y_test_pred)
train_evs = explained_variance_score(y_train, y_train_pred)
test_evs = explained_variance_score(y_test, y_test_pred)

# Affichage des résultats dans un tableau
results = pd.DataFrame({'Métrique': ['RMSE', 'MAE', 'MedAE', 'R2', 'EVS'], 
                        'Apprentissage': [train_rmse, train_mae, train_medae, train_r2, train_evs], 
                        'Test': [test_rmse, test_mae, test_medae, test_r2, test_evs]})
results.set_index('Métrique', inplace=True)
print(results)
```

### Interprètation

- Le modèle présente des performances cohérentes entre les données d'apprentissage et de test, ce qui indique une bonne capacité de généralisation.

- L'erreur quadratique moyenne (RMSE) est de `31459` sur l'apprentissage et de `32564` sur le test. Ces valeurs, relativement proches, montrent que le modèle n'est pas excessivement surajusté aux données d'entraînement. Une RMSE plus faible indique que les prédictions du modèle sont globalement proches des valeurs réelles, bien que des améliorations soient possibles.

- Le coefficient de détermination `R^2`, quant à lui, est de `0.84` sur l'apprentissage et de 0.83 sur le test. Cela signifie que le modèle explique environ `84 %` de la variance des données en apprentissage et `83 %` en test.

#### 2.1.4. Validation croisée

On tente ici d'améliorer la performance de l'arbre de régression en trouvant des paramètres optimaux avec une validation croisée. Pour ce faire, on utilise la méthode GridSearchCV pour trouver la profondeur optimale.

```python
from sklearn.model_selection import GridSearchCV

# Définition de la grille de paramètres
param_grid = {'max_depth': np.arange(1, 21)}

# Instanciation de l'estimateur de l'arbre de décision
tree = DecisionTreeRegressor(random_state=42)

# Instanciation de l'objet GridSearchCV
grid_tree = GridSearchCV(tree, param_grid, cv=5, scoring='neg_mean_squared_error')

# Entraînement de l'objet GridSearchCV
grid_tree.fit(X_train, y_train)

# Affichage de la profondeur optimale
print("Profondeur optimale :", grid_tree.best_params_['max_depth'])

# Affichage de la courbe de validation croisée pour la profondeur de l'arbre
plt.plot(param_grid['max_depth'], np.sqrt(-grid_tree.cv_results_['mean_test_score']))
plt.xlabel('Profondeur de l\'arbre')
plt.ylabel('RMSE')
plt.title('Validation croisée pour la profondeur de l\'arbre')
plt.show()
```

Commentaire : la courbe atteint son minimun à la pronfondeur 8 donc on entraîne de nouveau l'arbre avec cette nouvelle profondeur.

# V/ Conclusion Générale de l'Étude

L'objectif principal de cette étude était de comparer différentes approches de modélisation pour une tâche de régression, en évaluant leurs performances et leur capacité de généralisation sur un ensemble de données donné. Au cours de cette analyse, plusieurs modèles ont été explorés, notamment la régression linéaire multiple, l'arbre de décision, et la forêt aléatoire, avec et sans optimisation des hyperparamètres. Voici les principaux enseignements tirés de l'étude :

- **Régression linéaire multiple :**  
Ce modèle a montré des performances décentes, expliquant environ **83,4%** de la variance des données de test $( R^2 = 0.834 )$. Cependant, ses erreurs moyennes (RMSE et MSE) restent relativement élevées, suggérant qu'il ne capture pas efficacement les non-linéarités et les interactions complexes entre les variables.

- **Arbre de décision (profondeur limitée à 8) :**  
L'arbre de décision a eu des performances inférieures à celles de la régression linéaire $( R^2 = 0.711 )$, notamment en raison de sa faible profondeur, qui limite sa capacité à modéliser les relations complexes dans les données. Ce modèle est sensible au surapprentissage lorsqu'il est trop complexe ou sous-performe lorsqu'il est trop contraint.

- **Forêt Aléatoire :**  
Les modèles de Forêt Aléatoire, même sans optimisation, ont surpassé la régression linéaire et l'arbre de décision en termes de performance. Avec $( R^2 \approx 0.867 )$ et des erreurs réduites, ils ont démontré une excellente capacité de généralisation grâce à leur nature d'ensemble et leur robustesse face au bruit.

- **Optimisation avec GridSearchCV :**  
L'optimisation par **GridSearchCV** a produit les meilleurs résultats, atteignant un $( R^2 \approx 0.871 )$ et réduisant légèrement les erreurs (MSE). Toutefois, l'augmentation du nombre d'arbres $( n\_estimators=500 )$ a augmenté le coût computationnel.

- **Optimisation avec RandomizedSearchCV :**  
L'optimisation par **RandomizedSearchCV** a offert une alternative plus efficace. Avec $( R^2 \approx 0.867 )$, ses performances sont très proches de celles obtenues avec GridSearchCV, mais avec un **temps d'entraînement réduit** grâce à un nombre d'arbres réduit $( n\_estimators=200 )$ et une profondeur maximale fixée $max$ performances sur des échantillons variés.

---
Dans cette étude, la **Forêt Aléatoire** optimisée s'est imposée comme la solution la plus performante et robuste pour la tâche de régression, surpassant largement la régression linéaire et l'arbre de décision. Grâce à ses capacités d'ensemble, elle combine précision, robustesse, et généralisation efficace. L'optimisation des hyperparamètres via **RandomizedSearchCV** a confirmé qu'il est possible d'améliorer les performances tout en réduisant les coûts computationnels, en faisant de ce modèle un choix équilibré pour des applications pratiques.
